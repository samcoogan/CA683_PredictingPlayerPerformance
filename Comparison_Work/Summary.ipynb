```python
df_summary = pd.DataFrame(columns = ['MODEL','TRAIN_SIZE_IN_CV','TIME_PER_TRAIN','MAE','MSE','RMSLE','R2'],dtype = float)
for key in dict_result.keys():
  df = pd.read_csv(dict_result[key])
  record = pd.Series({\
    'MODEL':key[0],\
    'TRAIN_SIZE_IN_CV':key[1],\
    'TIME_PER_TRAIN':np.mean(df['TIME_PER_TRAIN']),\
    'MAE':np.mean(df['MAE']),\
    'MSE':np.mean(df['MSE']),\
    'RMSLE':np.mean(df['RMSLE']),\
    'R2':np.mean(df['R2']),\
    },name = str(key))
  df_summary = df_summary.append(record)
df_summary
```
