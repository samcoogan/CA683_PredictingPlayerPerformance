```python
dict_result = {
    ('LSTM',40):'./LSTM_CV_result40.csv',
    ('LSTM',80):'./LSTM_CV_result80.csv',
    ('GRU',40):'./GRU_CV_result40.csv',
    ('GRU',80):'./GRU_CV_result80.csv',
}
```

```python
df_summary = pd.DataFrame(columns = ['MODEL','TRAIN_SIZE_IN_CV','MAE','MSE','RMSLE','R2'],dtype = float)
for key in dict_result.keys():
  df = pd.read_csv(dict_result[key])
  record = pd.Series({\
    'MODEL':key[0],\
    'TRAIN_SIZE_IN_CV':key[1],\
    'MAE':np.mean(df['MAE']),\
    'MSE':np.mean(df['MSE']),\
    'RMSLE':np.mean(df['RMSLE']),\
    'R2':np.mean(df['R2']),\
    },name = str(key))
  df_summary = df_summary.append(record)
df_summary
```
